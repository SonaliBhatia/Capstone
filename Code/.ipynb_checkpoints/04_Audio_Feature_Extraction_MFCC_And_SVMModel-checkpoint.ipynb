```python
#Extract MFCC only
df = pd.DataFrame(columns=['MFCC'])
for i in range(data.shape[0]):
    filename=data.iloc[i].filename
    X, sample_rate = librosa.load(audio_files_dir+filename+'.mp3'
                                  ,res_type='kaiser_fast'
                                  ,sr=44100
                                  ,duration=10
                                 )
    sample_rate = np.array(sample_rate)
    mfccs = np.mean(librosa.feature.mfcc(y=X, 
                                        sr=sample_rate, 
                                        n_mfcc=13),
                    axis=0)
    df.loc[i]=[mfccs]
   # x,sr=librosa.load(audio_files_dir+filename+'.mp3')
   # mfcc = librosa.feature.mfcc(y=x, sr=sr)
```

# Extract multiple features -  this code was previously used for extracting multiple features. Leaving it here as example
#df=pd.DataFrame(columns=['MFCC','RMSE','Chroma_stft','Spec_cent','Spec_bw','Rolloff','ZCR'])

for i in range(data.shape[0]):

    filename=data.iloc[i].filename
    x,sr=librosa.load(audio_files_dir+filename+'.mp3')
    mfcc = librosa.feature.mfcc(y=x, sr=sr)
    rms = librosa.feature.rms(y=x)
    chroma_stft = librosa.feature.chroma_stft(y=x, sr=sr)
    spec_cent = librosa.feature.spectral_centroid(y=x, sr=sr)
    spec_bw = librosa.feature.spectral_bandwidth(y=x, sr=sr)
    rolloff = librosa.feature.spectral_rolloff(y=x, sr=sr)
    zcr = librosa.feature.zero_crossing_rate(x)
    df.loc[i]=[np.mean(mfcc),np.mean(chroma_stft), np.mean(rms), np.mean(spec_cent), np.mean(spec_bw), np.mean(rolloff), np.mean(zcr) ]

# Checking for basic NN for model

```python
new_data['Accent'].value_counts(normalize=True)
```

```python
new_data['Accent_num']=new_data['Accent'].map({'American':1 , 'Chinese':2 ,'Indian':3})
```

```python
new_data.head()
```

```python
#new_data.to_csv('../Data/Data_and_Features.csv',index=False)
```

```python
features=range(0,861)
```

```python
#pickle the DF to preserve Data Types to use in another notebook
new_data.to_pickle('Data_And_Features.pkl')
```

```python
X=new_data[features]
y=new_data['Accent_num']
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3,random_state=42,stratify=y )
```

```python
X_train.shape , X_test.shape
```

```python
ss=StandardScaler()
```

```python
X_train_sc=ss.fit_transform(X_train)
X_test_sc=ss.transform(X_test)
```

```python
X_train_sc.shape , X_test_sc.shape , y_train.shape , y_test.shape
```

```python
#X_train_sc = np.expand_dims(X_train_sc, axis=2)
#X_test_sc = np.expand_dims(X_test_sc, axis=2)
```

```python
X_train_sc.shape , X_test_sc.shape
```

```python
cnn=Sequential()
```

```python
#cnn.add(Conv1D(256, 8, padding='same',input_shape=(X_train_sc.shape[1],1)))
```

```python
cnn.add(Dense(256, activation='relu', input_shape=(X_train_sc.shape[1],)))

cnn.add(Dense(128, activation='relu'))
cnn.add(Dense(64, activation='relu'))
cnn.add(Dense(5, activation='softmax'))
```

```python
# compiler
cnn.compile(loss='sparse_categorical_crossentropy',
           optimizer = 'adam',
            metrics = ['accuracy']
           )
```

```python
history = cnn.fit(X_train_sc,
              y_train,
              batch_size=500,
              epochs=100,
              validation_data = (X_test_sc, y_test),
              verbose=1)
```

```python
train_acc = history.history['accuracy']
train_loss = history.history['loss']
test_loss = history.history['val_loss']
test_acc = history.history['val_accuracy']

df = pd.DataFrame(list(zip(train_acc, test_acc, train_loss, test_loss)), 
               columns =['train_acc', 'test_acc', 'train_loss', 'test_loss']) 

df['accuracy_diff'] = df['train_acc'] - df['test_acc']
df['loss_diff'] = df['train_loss'] - df['test_loss']

df.tail()
```

```python
train_loss = history.history['loss']
test_loss = history.history['val_loss']

# Set figure size.
plt.figure(figsize=(12, 8))

# Generate line plot of training, testing loss over epochs.
plt.plot(train_loss, label='Training Loss', color='#185fad')
plt.plot(test_loss, label='Testing Loss', color='orange')

# Set title
plt.title('Training and Testing Loss by Epoch', fontsize = 25)
plt.xlabel('Epoch', fontsize = 18)
plt.ylabel('sparse_categorical_crossentropy', fontsize = 18)
plt.xticks(np.arange(20), np.arange(20))

plt.legend(fontsize = 18);
```

```python
plt.plot(history.history['accuracy'], label='Train accuracy')
plt.plot(history.history['val_accuracy'], label='Test accuracy')
plt.legend();

```

```python
y_pred=cnn.predict(X_test)
```

```python
y_pred=np.argmax(y_pred,axis=1)
```

```python
cnf=confusion_matrix(y_test, y_pred)
```

```python
target_names=['American','Chinies','Indian']

print(classification_report(y_test, y_pred, 
 target_names=target_names))
```

```python
sns.heatmap(cnf,annot=True)
```
